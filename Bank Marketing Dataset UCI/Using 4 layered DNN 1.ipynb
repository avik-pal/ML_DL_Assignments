# PREPARE THE TRAINING AND TESTING DATA

```python
X = bank[features[0:-1]]
```

```python
Y = bank[features[-1]]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.80,stratify=Y)
```

# CREATING THE NEURAL NETWORK ARCHITECTURE

```python
from functools import partial
```

```python
X = tf.placeholder(dtype=tf.float32,shape=(None,16))
Y = tf.placeholder(dtype=tf.int32,shape=(None))
training = tf.placeholder_with_default(False,shape=(),name='training')
```

```python
my_batch_norm = partial(tf.layers.batch_normalization,training=training,momentum=0.99)
```

```python
hid1 = tf.layers.dense(X,units=1024)
bn1 = my_batch_norm(hid1)
bn1_act = tf.nn.elu(bn1)
layer1_out = tf.nn.dropout(bn1_act,keep_prob=0.7)
```

```python
hid2 = tf.layers.dense(layer1_out,units=512)
bn2 = my_batch_norm(hid2)
bn2_act = tf.nn.elu(bn2)
layer2_out = tf.nn.dropout(bn2_act,keep_prob=0.5)
```

```python
hid3 = tf.layers.dense(layer2_out,units=256)
bn3 = my_batch_norm(hid3)
bn3_act = tf.nn.elu(bn3)
layer3_out = tf.nn.dropout(bn3_act,keep_prob=0.5)
```

```python
hid4 = tf.layers.dense(layer3_out,units=128)
bn4 = my_batch_norm(hid4)
bn4_act = tf.nn.elu(bn4)
layer4_out = tf.nn.dropout(bn4_act,keep_prob=0.3)
```

```python
output_layer = tf.layers.dense(layer4_out,2,name='output')
```

```python
cost = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=output_layer,labels=Y)
```

```python
loss = tf.reduce_mean(cost,name='loss')
```

```python
training_op = tf.train.AdamOptimizer(learning_rate=0.01).minimize(loss)
```

```python
correct = tf.nn.in_top_k(output_layer, Y, 1)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

# TRAINING AND TESTING THE NEURAL NETWORK

```python
n_epochs = 50

init = tf.global_variables_initializer()
extra_update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
saver = tf.train.Saver()

epoch_list= []
training_accuracy = []
testing_accuracy = []
best_epoch = -1

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        sess.run([training_op,extra_update_ops], feed_dict={X:X_train,Y:Y_train,training:True})
        if(True):
            acc_train = accuracy.eval(feed_dict={X: X_train, Y: Y_train,training:True})
            acc_test = accuracy.eval(feed_dict={X: X_test,Y: Y_test})
            epoch_list.append(epoch)
            training_accuracy.append(acc_train)
            testing_accuracy.append(acc_test)
            print(epoch, "Train accuracy:", acc_train, "Test accuracy:", acc_test)
            if(best_epoch == -1):
                best_epoch = 0
                save_path = saver.save(sess, "/tmp/my_model.ckpt")
            elif(testing_accuracy[best_epoch]<testing_accuracy[-1]):
                best_epoch = epoch
                save_path = saver.save(sess, "/tmp/my_model.ckpt")
            elif(epoch-best_epoch >= 5 and epoch>=10):
                print("Best_epoch is ",best_epoch)
                break
```
